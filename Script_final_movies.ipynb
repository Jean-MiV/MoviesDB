```python
#Which are the highest Revenue generated Genres ?

data['Genre'].value_counts().head(15)
```

```python
# Created list to use for loop
lst = ['Action,Adventure,Sci-Fi',
        'Drama',
        'Comedy,Drama,Romance',
        'Comedy',
        'Drama,Romance',
        'Animation,Adventure,Comedy',
        'Comedy,Drama',
        'Action,Adventure,Fantasy',
        'Comedy,Romance',
        'Crime,Drama,Thriller',
        'Crime,Drama,Mystery',
        'Action,Adventure,Drama',
        'Action,Crime,Drama',
        'Horror,Thriller',
        'Drama,Thriller']


# Summation of revenue for each genre
TopGnr = []
for i in lst:
    x = int(data[data['Genre']==i]['Revenue (Millions)'].sum())
    TopGnr.append(x)

# TopGen DataFrame for the values.
TopGen= pd.DataFrame(TopGnr)
TopGen.columns =['Revenue_Sum(Millions)'] 
TopGen['Genres'] = lst

#TopGn.head()
```

```python
TopGen=TopGen.sort_values('Revenue_Sum(Millions)', ascending=True) #Sorting Values in ascending order for graph
```

```python
plt.figure(figsize=(9,9))
sns.barplot(x="Revenue_Sum(Millions)", y="Genres",data=TopGen,linewidth=1,ci= None)
```

```python
# Relation between Rating and Revenue.
plt.figure(figsize=(20,20))
sns.jointplot("Rating","Revenue (Millions)", data=data, kind='hex', color='DarkBlue')
#Relation between Rating and Revenue.
```

```python
#What is the optimal film runtime to achieve the best rating?
g = []
h = []
for i in range(4,10):
    den = data[data['Rating']==i]['Runtime (Minutes)'].count()
    g.append(int((data[data['Rating']==i]['Runtime (Minutes)'].sum())/den))
    h.append(i)
```

```python
dt1 = (pd.DataFrame(list(zip(g,h)), columns=['AvgRuntime','Rating'])) #.sort_values (ascending=False)
dt1
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x="Rating", y="AvgRuntime",data=dt1,ci= None)
```

```python
# What is the Revenue trend of the Movies over the Years?

x =[]
y =[]

for i in data['Year'].unique():
    y.append(data[data['Year']==i]['Revenue (Millions)'].sum())
    x.append(i) 
```

```python
dt2 = pd.DataFrame(list(zip(x, y)), columns=['Year','Revenue_Sum'])
dt2
```

```python
plt.figure(figsize=(10,7))
TrendRev = sns.lineplot(x="Year", y="Revenue_Sum",data=dt2,ci= None)
```
